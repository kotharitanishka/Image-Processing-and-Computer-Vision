```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
import random
```

```python
img = cv2.imread("horse.jpg" , 0)
img_con1 = cv2.imread("horse.jpg" , 0)
img_con2 = cv2.imread("horse.jpg" , 0)
img_lt = cv2.imread("horse.jpg" , 0)
img_pt = cv2.imread("horse.jpg" , 0)
plt.imshow(img , cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
height , width = img.shape
```

```python
#contrast stretching 
# l < 1 , m > 1 , n < 1

print("rmin before log trans = " , np.amin(img_con1))
print("rmax before log trans = " , np.amax(img_con1))

a = 60
b = 120
l = 0.4
n = 0.8
m = 1.5
v = l*a
w = v + m*(b-a)

for i in range(height):
    for j in range(width):
        if(0 < img_con1[i][j] <= a) :
            img_con1[i][j] = l*img_con1[i][j]
        elif(a < img_con1[i][j] <= b) :
            img_con1[i][j] = m*(img_con1[i][j] - a) + v
        elif(b < img_con1[i][j] <= 255):
            img_con1[i][j] = n*(img_con1[i][j] - b) + w

plt.imshow(img_con1 , cmap = "gray")
plt.axis('off')
plt.show()

print("rmin after log trans = " , np.amin(img_con1))
print("rmax after log trans = " , np.amax(img_con1))
```

```python
#contrast stretching 
# l > 1 , m < 1 , n > 1

print("rmin before log trans = " , np.amin(img_con2))
print("rmax before log trans = " , np.amax(img_con2))

a = 60
b = 120
l = 1.2
n = 1.8
m = 0.5
v = l*a
w = v + m*(b-a)

for i in range(height):
    for j in range(width):
        if(0 < img_con2[i][j] <= a) :
            img_con2[i][j] = l*img_con2[i][j]
        elif(a < img_con2[i][j] <= b) :
            img_con2[i][j] = m*(img_con2[i][j] - a) + v
        elif(b < img_con2[i][j] <= 255):
            img_con2[i][j] = n*(img_con2[i][j] - b) + w

plt.imshow(img_con2 , cmap = "gray")
plt.axis('off')
plt.show()

print("rmin after log trans = " , np.amin(img_con2))
print("rmax after log trans = " , np.amax(img_con2))
```

```python
#log transformation
print("rmin before log trans = " , np.amin(img_lt))
print("rmax before log trans = " , np.amax(img_lt))

c = 70

for i in range(height):
    for j in range(width):
        img_lt[i][j] = c * math.log(img_lt[i][j] + 1)

plt.imshow(img_lt , cmap = "gray")
plt.axis('off')
plt.show()

print("rmin after log trans = " , np.amin(img_lt))
print("rmax after log trans = " , np.amax(img_lt))
```

```python
#log transformation
print("rmin before log trans = " , np.amin(img_pt))
print("rmax before log trans = " , np.amax(img_pt))

y = 0.5

for i in range(height):
    for j in range(width):
        img_pt[i][j] = math.pow(img_pt[i][j] , y)

plt.imshow(img_pt , cmap = "gray")
plt.axis('off')
plt.show()

print("rmin after log trans = " , np.amin(img_pt))
print("rmax after log trans = " , np.amax(img_pt))
```
