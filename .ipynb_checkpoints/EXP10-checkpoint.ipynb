```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
import random
```

```python
img = cv2.imread('house.jpg', 0)
plt.imshow(img , cmap= 'gray')
plt.axis('off')
plt.show()
```

```python
h , w = img.shape
```

```python
#prewitt fx mask

p_fx_mask = np.array([[-1, -1, -1] , [0, 0, 0], [1, 1, 1]])
hp = np.zeros(img.shape)

for i in range(1, h-1):
    for j in range(1, w-1):
        temp = (img[i-1, j-1]*p_fx_mask[0][0] + img[i, j-1]*p_fx_mask[1][0] + img[i+1, j-1]*p_fx_mask[2][0] + 
                img[i-1, j]*p_fx_mask[0][1] +   img[i, j]*p_fx_mask[1][1] +   img[i+1, j]*p_fx_mask[2][1] +   
                img[i-1, j+1]*p_fx_mask[0][2] + img[i, j+1]*p_fx_mask[1][2] + img[i+1, j+1]*p_fx_mask[2][2] )
        hp[i][j] = temp
        
plt.imshow(hp, cmap = 'gray')
plt.axis('off')
plt.show()

#prewitt fy mask

p_fy_mask = np.array([[-1, 0, 1] , [-1, 0, 1], [-1, 0, 1]])
vp = np.zeros(img.shape)

for i in range(1, h-1):
    for j in range(1, w-1):
        temp = (img[i-1, j-1]*p_fy_mask[0][0] + img[i, j-1]*p_fy_mask[1][0] + img[i+1, j-1]*p_fy_mask[2][0] + 
                img[i-1, j]*p_fy_mask[0][1] +   img[i, j]*p_fy_mask[1][1] +   img[i+1, j]*p_fy_mask[2][1] +   
                img[i-1, j+1]*p_fy_mask[0][2] + img[i, j+1]*p_fy_mask[1][2] + img[i+1, j+1]*p_fy_mask[2][2] )
        vp[i][j] = temp
        
plt.imshow(vp, cmap = 'gray')
plt.axis('off')
plt.show()

#prewitt fx + fy mask
p = np.zeros(img.shape)

for i in range(1, h-1):
    for j in range(1, w-1):
        
        temp_x = (img[i-1, j-1]*p_fx_mask[0][0] + img[i, j-1]*p_fx_mask[1][0] + img[i+1, j-1]*p_fx_mask[2][0] + 
                img[i-1, j]*p_fx_mask[0][1] +   img[i, j]*p_fx_mask[1][1] +   img[i+1, j]*p_fx_mask[2][1] +   
                img[i-1, j+1]*p_fx_mask[0][2] + img[i, j+1]*p_fx_mask[1][2] + img[i+1, j+1]*p_fx_mask[2][2] )

        temp_y = (img[i-1, j-1]*p_fy_mask[0][0] + img[i, j-1]*p_fy_mask[1][0] + img[i+1, j-1]*p_fy_mask[2][0] + 
                img[i-1, j]*p_fy_mask[0][1] +   img[i, j]*p_fy_mask[1][1] +   img[i+1, j]*p_fy_mask[2][1] +   
                img[i-1, j+1]*p_fy_mask[0][2] + img[i, j+1]*p_fy_mask[1][2] + img[i+1, j+1]*p_fy_mask[2][2] )
        
        temp = np.sqrt((pow(temp_x , 2)) + (pow(temp_y , 2)))
        p[i][j] = temp

plt.imshow(p, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
#sobel fx mask

s_fx_mask = np.array([[-1, -2, -1] , [0, 0, 0], [1, 2, 1]])
hs = np.zeros(img.shape)

for i in range(1, h-1):
    for j in range(1, w-1):
        temp = (img[i-1, j-1]*s_fx_mask[0][0] + img[i, j-1]*s_fx_mask[1][0] + img[i+1, j-1]*s_fx_mask[2][0] + 
                img[i-1, j]*s_fx_mask[0][1] +   img[i, j]*s_fx_mask[1][1] +   img[i+1, j]*s_fx_mask[2][1] +   
                img[i-1, j+1]*s_fx_mask[0][2] + img[i, j+1]*s_fx_mask[1][2] + img[i+1, j+1]*s_fx_mask[2][2] )
        hs[i][j] = temp
        
plt.imshow(hs, cmap = 'gray')
plt.axis('off')
plt.show()

#sobel fy mask

s_fy_mask = np.array([[-1, 0, 1] , [-2, 0, 2], [-1, 0, 1]])
vs = np.zeros(img.shape)

for i in range(1, h-1):
    for j in range(1, w-1):
        temp = (img[i-1, j-1]*s_fy_mask[0][0] + img[i, j-1]*s_fy_mask[1][0] + img[i+1, j-1]*s_fy_mask[2][0] + 
                img[i-1, j]*s_fy_mask[0][1] +   img[i, j]*s_fy_mask[1][1] +   img[i+1, j]*s_fy_mask[2][1] +   
                img[i-1, j+1]*s_fy_mask[0][2] + img[i, j+1]*s_fy_mask[1][2] + img[i+1, j+1]*s_fy_mask[2][2] )
        vs[i][j] = temp
        
plt.imshow(vs, cmap = 'gray')
plt.axis('off')
plt.show()

#sobel fx + fy mask
s = np.zeros(img.shape)

for i in range(1, h-1):
    for j in range(1, w-1):
        
        temp_x = (img[i-1, j-1]*s_fx_mask[0][0] + img[i, j-1]*s_fx_mask[1][0] + img[i+1, j-1]*s_fx_mask[2][0] + 
                img[i-1, j]*s_fx_mask[0][1] +   img[i, j]*s_fx_mask[1][1] +   img[i+1, j]*s_fx_mask[2][1] +   
                img[i-1, j+1]*s_fx_mask[0][2] + img[i, j+1]*s_fx_mask[1][2] + img[i+1, j+1]*s_fx_mask[2][2] )

        temp_y = (img[i-1, j-1]*s_fy_mask[0][0] + img[i, j-1]*s_fy_mask[1][0] + img[i+1, j-1]*s_fy_mask[2][0] + 
                img[i-1, j]*s_fy_mask[0][1] +   img[i, j]*s_fy_mask[1][1] +   img[i+1, j]*s_fy_mask[2][1] +   
                img[i-1, j+1]*s_fy_mask[0][2] + img[i, j+1]*s_fy_mask[1][2] + img[i+1, j+1]*s_fy_mask[2][2] )
        
        temp = np.sqrt((pow(temp_x , 2)) + (pow(temp_y , 2)))
        s[i][j] = temp

plt.imshow(s, cmap = 'gray')
plt.axis('off')
plt.show()
```
