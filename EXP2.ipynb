```python
import cv2
import matplotlib.pyplot as plt
import math
import random
```

```python
img = cv2.imread("house.jpg" , 0)
img_neg = cv2.imread("house.jpg" , 0)
img_th = cv2.imread("house.jpg" , 0)
img_gs_wb = cv2.imread("house.jpg" , 0)
img_gs_wob = cv2.imread("house.jpg" , 0)
plt.imshow(img , cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
h,w = img.shape
print(h)
print(w)
```

```python
#img negative
for i in range(h):
    for j in range(w):
            img_neg[i][j] = 255 - img_neg[i][j]
plt.imshow(img_neg , cmap= "gray")
plt.axis('off')
plt.show()
```

```python
#thresholding
a = 120
for i in range(h):
    for j in range(w):
        if (img_th[i][j] < a):
            img_th[i][j] = 0
        else :
            img_th[i][j] = 255
plt.imshow(img_th , cmap= "gray")
plt.axis('off')
plt.show()
```

```python
#gray level slicing with background
a = 100
b = 150
for i in range(h):
    for j in range(w):
        if ( a <= img_gs_wb[i][j] <= b):
            img_gs_wb[i][j] = 255
plt.imshow(img_gs_wb , cmap= "gray")
plt.axis('off')
plt.show()
```

```python
#gray level slicing without background
a = 100
b = 150
for i in range(h):
    for j in range(w):
        if ( a <= img_gs_wob[i][j] <= b):
            img_gs_wob[i][j] = 255
        else : 
            img_gs_wob[i][j] = 0
plt.imshow(img_gs_wob , cmap= "gray")
plt.axis('off')
plt.show()
```
