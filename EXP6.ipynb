```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
img = cv2.imread("horse.jpg" , 0)
plt.imshow(img , cmap = "gray")
plt.axis('off')
plt.show()
```

```python
h ,  w = img.shape
```

```python
#histogram stretching
rmin = img.min()
rmax = img.max()
smax = 255
smin = 0

hs_img = np.zeros(img.shape)

for i in range(h):
    for j in range(w):
        r = img[i][j]
        s = (((smax - smin) / (rmax - rmin)) * (r - rmin)) + smin
        hs_img[i][j] = s
        
print("rmin of original image is : " , img.min())
print("rmax of original image is : " , img.max())
print("\nrmin of new image is : " , hs_img.min())
print("rmax of new image is : " , hs_img.max())    
plt.imshow(hs_img , cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
#count the frequency
def get_freq(img):
    r = []
    count = []
    
    for k in range(0,256):
        r.append(k)
        c = np.count_nonzero(k == img)
        count.append(c)
    
    return(r,count)
```

```python
#comparing histogram of original and histogram stretched images 
fig = plt.figure(figsize =(15,10))
rows = 2
cols = 1

fig.add_subplot(rows, cols , 1)
r, count = get_freq(img)
plt.stem(r , count)
plt.xlabel("intensity value")
plt.ylabel("no of pixels")
plt.title("histogram of original image")

fig.add_subplot(rows, cols , 2)
r, count = get_freq(hs_img)
plt.stem(r ,count)
plt.xlabel("intensity value")
plt.ylabel("no of pixels")
plt.title("histogram of new image")

plt.show()
```

```python
#comparing original and histogram stretched images

fig = plt.figure(figsize =(15,5))
rows = 1
cols = 2

fig.add_subplot(rows, cols , 1)
plt.imshow(img , cmap = "gray")
plt.axis("off")
plt.title("original img")

fig.add_subplot(rows, cols , 2)
plt.imshow(hs_img , cmap = "gray")
plt.axis("off")
plt.title("new img")

plt.show()
```
